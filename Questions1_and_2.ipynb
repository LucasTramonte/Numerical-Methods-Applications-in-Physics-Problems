```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def a(t): #defining the acceleration function
  return 75*(-3/2 +t)/2000
t = np.arange(0.0, 5.0, 0.01) #discretizing time
plt.figure()
plt.subplot(211) 
plt.grid()
plt.plot(t,a(t),'k', lw = 2., label='acceleration(m/s^2)') 

#Creating the speed graph :
plt.grid()
plt.plot(t, 75*(-3+t)*t/4000,'r',lw = 2., label='velocity(m/s)') # Another possibility is to define the velocity function, as was done for acceleration:

#Creating the position graph :
plt.grid()
plt.plot(t, 75*(-3/2 +t/3)*t*t/4000,'b',lw = 2., label='position(m)') 
plt.xlabel('time (s)', fontsize=12)
plt.legend()
plt.show()
```

### Finding the distance traveled in 5 seconds

```python
t = np.arange(0.0, 5.0, 0.005)
plt.figure()
plt.subplot(211)
plt.grid()
plt.title("Velocity modulus")
def v(t) :
  return abs(75*(-3+t)*t/4000) # Velocity modulus 
plt.plot(t, v(t),'r',lw = 2., label='Velocity(m/s)')
plt.xlabel('time (s)', fontsize=12)
plt.legend()
plt.show()

dt = 0.001
def s(t): #defining the distance traveled
  result = 0 
  for i in range(0, 5000):
    result = result + dt*v(i*dt)
  return result

print("\n Therefore, the distance covered in 5 seconds is :" , s(t), "m")
```

```python
L = 1.2    # rope length (m)
g = 9.81   # acceleration of gravity
v = 5      # initial velocity

# determine theta solving  cos^2(teta) + ((v^2)/(g*L))*cos(teta) - 1 = 0

a = 1
b = (np.power(v, 2))/(g*L)
c = -1

def raizes(a, b, c):
  Delta = (b**2 - 4*a*c)
  x1 = (-b + Delta**(1/2)) / (2*a)
  x2 = (-b - Delta**(1/2)) / (2*a) # not suitable
  return x1

teta1 = math.acos(raizes(a,b,c))

print('The angle of the conical pendulum is :',180*teta1/(np.pi), 'degrees')

def d(teta): #defining the displacement of the horizontal throw
  return  L*np.sqrt(2)*np.sin(teta)*np.sqrt(2/(np.cos(teta)) -1)

def a(teta): #defining the radius a
  return  ((L*np.sin(teta))**(2) + d(teta)**(2))**(1/2)

print('a = ', a(teta1), 'm')

teta = np.arange(0.0, 1.7, 0.01) #discretizing the angle
plt.figure()
plt.subplot(211) 
plt.grid()
plt.axis([0, 1.7, 0, 20]) # defining the x-axis and y-axis intervals
plt.plot(teta,a(teta),'k', lw = 2., label='a (m)') 
plt.xlabel('teta (rad)', fontsize=12)
plt.ylabel('a (m)',fontsize=12)
plt.axvline((np.pi)/2, 0, 4, color = 'r', label='90 graus')

plt.plot(teta1, a(teta1), 'bo')
plt.legend()

plt.show()
```
